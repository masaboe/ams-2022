## Topic Modelling Menggunakan Latent Dirchlect Allocation

##### Apa itu topic modelling? 
Secara sederhana adalah mengelompokan data text berdasarkan suatu topik tertentu. 

<b>Bagaimana dengan klasifikasi? </b>

Memiliki tujuan yang sama dengan klasifikasi tetapi menggunakan pendekatan berbeda. Topik modelling merupakan <b>unsupervised learning</b> (tidak membutuhkan data berlabel). bisa dikatakan topic modelling bekerja seperti clustering dengan mengelompokan dokumen berdasarkan kemiripanya, tetapi topic modelling mempunyai tujuan yang lebih spesifik

##### Tujuan Topik Modelling
- Menemukan pola topik abstrak pada kumpulan dokumen
- Memberikan anotasi dokumen berdasarkan topik tersebut
- Menggunakan Anotasi dokumen untuk mengelompokan dokumen

##### Bagaimana LDA Bekerja?
Lda bekerja dengan menggunakan <b>gibbs sampling</b>. 

Gibbs sampling adalah salah satu algoritma keluarga dari Markov Chain Monte Carlo (MCMC). Kurang lebih intinya adalah  bisa melakukan perhitungan joint probability distribution dengan cara melakukan sampling satu per satu terhadap setiap variabel dengan berdasarkan nilai variabel lainnya alias full conditional probability.

Dalam melakukan pengelompokan topik ada dua bentuk distribusi probabilitas yang harus dicari yaitu :
- Distribusi Probabilitas dokumen pada suatu dokumen
- Distribusi Probabilitas kata pada suatu topik

#### Distribusi Probabilitas topik pada suatu dokumen
Probabilitas topik pada dokumen merupakan nilai probabilitas tiap topik pada suatu dokumen. misal pada dokumen I mempunyai probabilitas topik A senilai X, mempunyai probabilitas topik B senilai Y dan seterusnya sesuai dengan jumlah topik. nilai tersebut dapat kita temukan melalui rumus.



Dimana : K merupakan indeks Topik dan D merupakan indeks Dokumen dan alpha merupakan parameter dirchlect.

Parameter  digunakan dalam menentukan distribusi topik dalam dokumen, semakin besar nilai alpha dalam suatu dokumen, menandakan campuran topik yang dibahas dalam dokumen semakin banyak.

#### Distribusi Probabilitas kata pada suatu topik

probabilitas kata pada topikmerupakan nilai probabilitas tiap kata pada suatu topik. misal pada kata I mempunyai probabilitas topik A senilai X, mempunyai probabilitas topik B senilai Y dan seterusnya sesuai dengan jumlah topik. nilai tersebut dapat kita temukan melalui rumus



Dimana : t merupakan indeks token dan K merupakan indeks topik dan beta merupakan parameter dirchlect.

Parameter  digunakan untuk menentukan distribusi kata dalam topik. Semakin tinggi nilai beta, maka semakin banyak kata-kata yang ada di dalam topik, sedangkan semakin kecil nilai beta, maka semakin sedikit kata-kata yang ada di dalam topik sehingga topik tersebut mengandung kata-kata yang lebih spesifik

#### Mengupdate topik pada suatu kata

jika sudah memiliki kedua nilai distribusi diatas maka kita bisa mengupdate topik pada tiap kata. dengan cara melakukan sample dengan bobot dimana bobot tersebut didapatkan dari nilai Distribusi Probabilitas topik pada suatu dokumen dikali Distribusi Probabilitas kata pada suatu topik



Setelah iterasi berhasil kita akan menentukan topik pada suatu dokumen dengan melabeli dokumen tersebut dengan probabilitas topik paling besar. dan kita juga akan mendapatkan annotation word dari tiap topik seperti gambar paling atas.
